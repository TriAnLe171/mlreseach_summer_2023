```python
import math

# this class should interpret smth similar to the node in a normal tree
class Position:
    def __init__(self, children, value):
        self.children = children
        self.value = value

    def getChildren(self):
        return self.children

    def value(self):
        return self.value


# the whole algorithm should have a flag that indicates if the game is over
game_over = False

def minimax(position, depth, alpha, betta, maximizingPlayer):
    if depth == 0 or game_over:
        return None
        # but hsould return a static evaluation of position
    if maximizingPlayer:
        maxEval = -(math.inf)
        for child in position.getChildren():
            val = minimax(child, depth=depth-1, alpha=alpha, betta=betta, maximizingPlayer=False)
            maxEval = math.max(maxEval, val)
            alpha = math.max(alpha, val)
            if betta <= alpha:
                break
        return maxEval

    else:
        minEval = math.inf
        for child in position.getChildren():
            eval = minimax(child, depth=depth-1, alpha=alpha, betta=betta, maximizingPlayer=True)
            minEval = math.min(minEval, eval)
            betta = math.min(minEval, betta)
            if betta <= alpha:
                break
        return minEval



```
